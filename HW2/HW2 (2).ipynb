a) Aggregate the data by decade and count the number of expeditions, the total number of expedition members, and the average rates of death for members and hired staff. Briefly describe what you found.

```python
exp=expeditions.copy()
```

```python
exp["decade"]=exp["year"]//10*10
```

```python
exp[exp["decade"]==1920]
```

```python
t_0=exp.groupby("decade")["members"].sum()
t_1=exp.groupby("decade")["expedition_id"].count()
t_2=exp.groupby("decade")["member_deaths"].sum()
t_3=exp.groupby("decade")["hired_staff_deaths"].sum()
t_4=exp.groupby("decade")["hired_staff"].sum()
```

```python
t_2=t_2/t_0
t_3=t_3/t_4
t_2=pd.DataFrame(t_2)
t_3=pd.DataFrame(t_3)
t_2
```

```python
t=pd.merge(t_0,t_1,on="decade")
tt=pd.merge(t_2,t_3,on="decade")
t=pd.merge(t,tt,on="decade")
```

```python
t=t.rename(columns={"0_x":"member death ratio"})
t=t.rename(columns={"0_y":"hired staff death ratio"})
```

```python
t
```

The death ratio for members and hired staff decreases as time goes by and the number of members and expeditions increases gradually. 

b) Calculate the length_of_expedition as the time between the basecamp_date and the termination_date in days. Now, standardize these length_of_expedition values to z-values grouped by each peak (that is you need to standardize with the mean and standard deviation of expeditions for the same peak only). Are longer expeditions more or less likely to be associated with death? (no statistical test necessary)

```python
from datetime import date
exp[["basecamp_date","termination_date"]] = exp[["basecamp_date","termination_date"]].apply(pd.to_datetime) 
exp["length_of_expedition"]=(exp["termination_date"]-exp["basecamp_date"]).dt.days
```

```python
s=exp["length_of_expedition"].std()
```

```python
m=exp["length_of_expedition"].mean()
```

```python
exp["z_value"]=(exp["length_of_expedition"]-m)/s
```

```python
l=pd.DataFrame(exp.groupby("peak_name")["z_value"].mean())
```

```python
l
```

```python
b=exp.groupby("peak_name")["length_of_expedition"].mean()
```

```python
v=exp.groupby("peak_name")["member_deaths"].sum()
```

```python
pd.merge(b,v,on="peak_name")
```

From the data we got, it seem longer expeditions may lead higher death ratio.
