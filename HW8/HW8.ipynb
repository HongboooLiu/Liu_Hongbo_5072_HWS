5. API Client

a) API client function

```python
def steam_game_data(APIkey,user_id):
    # user_id is the 17 digital user id which can be found by every steam user in their account
    # APIkey is the APIkey provided by steam, register through https://steamcommunity.com/dev/registerkey
    # The output is a dataframe describe the games own by the input user and the playtime_fovrever indicates
    # the time played by the user and rtime_last_played indicates when this game was last played
    # Example:
    # input:
    # steam_game_data(APIkey,76561198911864850)
    # output:
    # Success
    # steam_df
    params = {'key':APIkey, 'steamid': str(user_id) ,'format':'json'}
    response = requests.get('http://api.steampowered.com/IPlayerService/GetOwnedGames/v0001/', params = params)
    if response.status_code== 404:
        print("error (failed to make request)")
    elif response.status_code== 500:
        print("successfully made request but had internal error")
    else:
        print("Success")
        steam = response.json()
        steam_df = pd.DataFrame(steam['response']['games'])
        return steam_df
```
